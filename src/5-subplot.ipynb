### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Validate input
if len(categories) != len(values):
    raise ValueError("Categories and values must have the same length.")
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes must have the same length.")
if sum(sizes) != 100:
    raise ValueError("The sum of sizes must be 100 for the pie chart.")

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Plot')

axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Plot')

axs[1, 0].scatter(x, y, color='purple')
axs[1, 0].set_title('Scatter Plot')

axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```
