### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

if len(x) != len(y):
    raise ValueError("x and y lists must have the same length.")
if not x or not y:
    raise ValueError("x and y lists cannot be empty.")

plt.scatter(x, y, color='blue')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

plt.title('Scatter Plot with Annotations')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```
