### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

x = list(range(11))
y = [i**2 for i in x]

if len(x) != len(y):
    raise ValueError("x and y must have the same length.")

plt.plot(x, y, marker='o', linestyle='-', color='b')


plt.title('Line Plot of x vs xÂ²')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)

plt.show()
```
